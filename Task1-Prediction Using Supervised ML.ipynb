## Simple Linear Regression
 we will predict the percentage of marks that a student is expected to score based upon the number of hours they studied in our task. As we know Two variables involved in this we will utilise these variables.
 

### Importing All the Required Libraries in Task


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Reading Data From Dataset

```python
gk=pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
print("Data imported ")
gk.head(15)

```

```python
gk.tail()
```

### Plotting the Distribution of Scores

```python
gk.plot(x='Hours', y='Scores',style='+',color='Green')
plt.title('Hours vs Percentage',size='30')  
plt.xlabel('Hours Studied',size='20')  
plt.ylabel('Percentage Score',size='20')  
plt.show()
```

we can clearly see the linear relationship between the variables No of Hours Studied and Percentage Score in the above graph.

```python
gk.describe()
```

```python
gk.info()
```

### Data preparation

```python
#Devide data into Attributes and Labels
X = gk.iloc[:, :-1].values  
y = gk.iloc[:, 1].values  
```

the next step is to split this data into training and test sets

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state=0) 
#we split our data in Training and Testing set.
```

### Training Algorithm
Now we have to train our Algorithm

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 
print("Algorithm Trained")
```

### Plotting the Regression Line and Test data

```python
line = regressor.coef_*X+regressor.intercept_
plt.scatter(X, y)
plt.plot(X, line,color='Red');
plt.show()
```

### Prediction
Algorithm Trained Successfully,its time to Predict.

```python
print(X_test) # Testing data In Hours
y_pred = regressor.predict(X_test) # Predicting the scores
```

```python
# Comparing Actual vs Predicted
gf = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
gf
```

### Testing With our Data

```python
hours = 9.25
test=np.array([hours])
test=test.reshape(-1,1)
own_pred = regressor.predict(test)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

### Evaluation Of Model
This is last step to evaluate the performance of Algorithm. We are using Mean Absolute Error in our Model.

```python
from sklearn import metrics
print('Mean absolute Error:',metrics.mean_absolute_error(y_test, y_pred)) 
```

If a student Studies 9.25Hours Per day, The model Predicts the score of 93.691%
